```python
import pandas as pd

# 加载数据
file_path = R'data\多元回归分析演示数据.xlsx'
数据表 = pd.read_excel(file_path)

# 显示数据的前几行以进行初步检查
数据表.head()
```

```python
import statsmodels.api as sm
# 使用指定的五个变量作为自变量进行多元线性回归分析
specified_variables = ['知识获取', '内部运作', '适应学习', '学习成长', '知识创新']
X_specified = 数据表[specified_variables]
y = 数据表['组织效能']  # 因变量

# 多元线性回归
X_specified_with_constant = sm.add_constant(X_specified)  
model_specified = sm.OLS(y, X_specified_with_constant).fit()

# 显示回归结果
model_specified_summary = model_specified.summary()
model_specified_summary
```

```python
import statsmodels.api as sm
# 使用指定的五个变量作为自变量进行多元线性回归分析
specified_variables = ['福利措施', '内部运作', '财务控管', '知识创新', '顾客认同','知识流通','同侪关系','适应学习','创新学习','知识获取','组织效能']
X_specified = 数据表[specified_variables]
y = 数据表['内部运作']  # 因变量

# 多元线性回归
X_specified_with_constant = sm.add_constant(X_specified)  
model_specified = sm.OLS(y, X_specified_with_constant).fit()

# 显示回归结果
model_specified_summary = model_specified.summary()
model_specified_summary
```
